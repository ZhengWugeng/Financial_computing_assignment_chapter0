### 2 Numerical Computing (Week 2)

##### 2.1 Newtonâ€™s Method


```python
def myfunc(x):
    y = (1 + 0.5 * x)**5
    return y
```

```python
def numerical_grad(func, x, dx=0.00001):
    dy = func(x + dx) - func(x)
    ngrad = dy / dx
    return ngrad
```

```python
def newton_method(func,
                  func_value=1.1,
                  x=-1,
                  max_iteration=1000000,
                  max_err=0.000001):
    for trial in range(max_iteration):
        error = func(x) - func_value
        if abs(error) < max_err:
            y = func(x)
            print(" Iteration {}: f = {}, x = {}".format(trial, y, x))
            return x
        else:
            grad = numerical_grad(func,x)
            x = x - error / grad
            y = func(x)
            print(" Iteration {}: f = {}, x = {}".format(trial, y, x))
    raise ValueError("Max iteration reached ")
    print("Max Iteration {}: f = {}, x = {}".format(trial, y, x))
```

```python
newton_method(myfunc)
```

##### 2.2 Gradient Descent


```python
def the_function(x):
    y = x**2 - 2 * x + 3
    return y
```

There is a problem: the grad is so large that the abs(df) cannot smaller than 0.000001
If I set grad is 0.005, the program can calculate the result

```python
def Gra_func(func, x=3, max_iteration=1000000, max_err=0.000001):
    d = 0.01
    # grad = numerical_grad(func,x)
    grad = 0.005

    for trial in range(max_iteration):
        x_new = x - d * grad
        df = func(x_new) - func(x)
        if (abs(df) < max_err):
            print(" Iteration {}: f = {}, x = {}".format(trial, df, x))
            return x
        else:
            x = x_new
            print(" Iteration {}: f = {}, x = {}".format(trial, df, x))
    raise ValueError("Max iteration reached ")
    print("Max Iteration {}: f = {}, x = {}".format(trial, df, x))
```

```python
Gra_func(the_function)
```
